##### This code shows how to do one hot encoding from scratch on a pandas dataframe.

```python
import pandas as pd
#initializing variables which I will use to create a toy dataset
a='a'
b='b'
c='c'
d='d'
e='e'
f='f'
g='g'
#creating a toy dataset using pandas and dictionaries
data=pd.DataFrame({'region':[a,b,a,c,a,b,c,a,a,b,c,a,b,c,e],'village':[d,f,d,g,e,e,f,g,d,f,e,d,f,g,e]})
col=data.columns #collecting the columns form the dataframe 
one_hot={} # This dictionary is going to hold the new one hot encoded columns
for i in col:
    categories=data[i].unique() #picking the categories of a parcular column. In this case column i
    for j in categories:
        new="{}_{}".format(i,j) #Creating a new column title
        lis=[] #This list will contain the values for the new column
        for k in data[i]:#This loop does the one hot encoding
            if k==j:
                lis.append(1)
            else:
                lis.append(0)
            one_hot[new]=lis
encoded_data_frame=pd.DataFrame(one_hot)
encoded_data_frame
```
